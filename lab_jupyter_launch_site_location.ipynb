```python
# Task 2: Mark the success/failed launches for each site on the map


# Example DataFrame with manually added success/failure data
launch_sites = pd.DataFrame({
    'Launch Site': ['CCAFS SLC 40', 'VAFB SLC 4E', 'KSC LC 39A', 'CCAFS LC 40'],
    'Latitude': [28.5623, 34.632834, 28.573255, 28.561857],
    'Longitude': [-80.577366, -120.610746, -80.646895, -80.577366],
    'Outcome': [1, 0, 1, 0]  # 1 indicates success, 0 indicates failure
})

# NASA's Kennedy Space Center coordinate for initializing the map
nasa_coordinate = [28.573255, -80.646895]

# Initialize the map centered around the average location of the launch sites
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterate through the DataFrame and add circles and markers
for index, site in launch_sites.iterrows():
    # Define the color based on the success (green) or failure (red) of the launch
    color = 'green' if site['Outcome'] == 1 else 'red'
    
    # Create a Circle object with a specific radius and color based on success/failure
    folium.Circle(
        location=[site['Latitude'], site['Longitude']],
        radius=1000,  # Radius in meters
        color=color,  # Circle border color based on success/failure
        fill=True,
        fill_color=color  # Circle fill color based on success/failure
    ).add_child(folium.Popup(site['Launch Site'])).add_to(site_map)
    
    # Create a Marker object with a label and color based on success/failure
    folium.Marker(
        location=[site['Latitude'], site['Longitude']],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:{};"><b>{}</b></div>'.format(color, site['Launch Site']),
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
